Reverse a number and check palindrom 
palindrom = 151<==>151 , string me intiger ko conv mt krna

```python
#REVERSE
n = 1234
num = n
rev= 0
while num>0:
    ld = num%10
    rev= (rev*10)+ld
    num = num//10
print(rev)
```

```python
#n = int(input("enter your number"))
n=12321
num = n
rev = 0
while num>0:
    ld = num%10
    rev = (rev*10)+ld
    num = num//10
if rev == n:
    print("palindrome")
else:
    print("not palindrome")
```

Armstrong Number
uske har digit ka power (number of digits ke barabar) ka sum usi number ke equal hota hai

EX. 153---> 1^3+3^3+5^3=153 armstrong

```python
# firstly sum of digits
n = 153
num = n
res = 0
while num>0:
    ld= num%10
    res= res+ld
    num = num//10
print(res)
```

```python
# secondly sum of cubes
n = 153
num = n
res = 0
while num>0:
    ld = num%10
    res = res+(ld**2)  # square
    num = num//10
print(res)
```

```python
# ARMSTRTONG CHECKING (without len function)
n = 153
num = n
sum = 0
digit=0
while num>0:
    digit+=1
    num = num//10 # is loop me digit count kari hai
num = n
sum = 0
while num>0: # is loop me sum
    ld = num%10
    sum = sum+ld**digit
    num = num // 10
if sum==n:
    print("Armstrong")
else:
    print("not Armstrong")
```

```python
# with len functions
n = 153
num = n
sum = 0
nod = len(str(n))
while num>0:
    ld = num%10
    sum = sum+(ld**nod)
    num = num//10
if sum==n:
    print("Armstrong")
else:
    print("not Armstrong")

# Time complexity ---> O(log10n)
# Space Complexity -->O(1)
```

Print factors / Divisors
- 10-->{1 2 5 10}
- 15-->{1 3 5 15}
- 25-->{1 5 25}

```python
# Brute Force
n = 10
num = n
count=[]
for i in range(1,num+1):
    if num%i==0:
        count.append(i)
print(count)
 # Time complexity O(n)
 # Space complexity o(k)--> k is no of solutions
```

```python
# Better solution
n = 10
num = n
count=[]
for i in range(1,num//2):  # run till half O(N/2)
    if num%i==0:
        count.append(i)
count.append(num) #--->O(1)
print(count)

# Time complexity O(N)
# Space complexity 0(k)
```

```python
# Optimal solution
from math import sqrt
n = 36
num = n
result = []
for i in range(1,int(sqrt(num))+1): # loop n-1 tk chalta hai 
    if num%i==0:
        result.append(i)
        if num//i !=i:
            result.append(num//i)
result.sort()
print(result)

# Time complexity(O\/n)+o(nlogn)
# Space Complexity O(k)

```
