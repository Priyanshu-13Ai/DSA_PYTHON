*What is Recursion?*

Recursion tab hoti hai jab ek function apne aap ko hi call karta hai
jab tak condition false na ho jaye.

Recursion = Function calling itself to solve a smaller version of the same problem

______________________________________________________________________________________

Structure of Recursion 

Har recursion me 2 cheezein compulsory hoti hain 

1️. Base Case

Jaha recursion rukta hai

Agar base case nahi → infinite loop 

2️. Recursive Call

Function khud ko smaller problem ke saath call karta hai

_________________________________________________________________________________________

Types of Recursion

1. Direct Recursion

Function directly khud ko call karta hai

EX. def fun():
    print("Hello")
    fun()

2. Indirect Recursion

Ek function doosre function ko call kare
aur wo phir pehle function ko call kare

EX. def fun1():
    fun2()

def fun2():
    fun1()

3. Tail Recursion

Recursive call last statement hoti hai

EX. def fact(n, ans=1):
    if n == 0:
        return ans
    return fact(n-1, ans*n)
Print pehle hota hai

4. Head Recursion

Recursive call pehle hoti hai, baad me kaam

EX. def print_num(n):
    if n == 0:
        return
    print_num(n-1)
    print(n)
print bad me hota hai

5. Tree Recursion

Function multiple recursive calls karta hai

EX. def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

6. Nested Recursion

Recursive call ke andar recursive call

EX. def fun(n):
    if n > 100:
        return n
    return fun(fun(n+11))

```python
# 1. print numbers n to 1
def number(n):
    if n == 0:          # base case
        return           #Jab n = 0 ho jaye → function ruk jaye 
                         #Agar ye nahi likhte ❌ → infinite recursion → error
    print(n)
    number(n-1)         # recursive call
number(5)
```

```python
#2. Factorial
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    # Recursive case
    return n * factorial(n - 1)
# Function call
print(factorial(5))
```

```python
#3. Sum of digits
def sum(n):
    if n==0:
        return 0
    return (n%10)+sum(n//10) # sum(n//10) me function call ho raha hai
sum(123)
"""
n // 10 → last digit hata deta hai

sum(n // 10) → baaki digits ka sum nikaalne ke liye khud ko call

            sum(123)
            = (123 % 10) + sum(12)
            = 3 + sum(12)
"""
```

```python
#5. Fibonacci (recursion + DP)
"""

Fibonacci series ek number series hoti hai jisme:

Har number = pichhle do numbers ka sum

"""
def fib(n):
    # Base case
    if n == 0 or n == 1:
        return n
    
    # Recursive case
    return fib(n-1) + fib(n-2)


print(fib(5))
```

```python
# dynamic programing(will we done next)
def fib_dp(n, dp):
    if n == 0 or n == 1:
        return n
    
    if dp[n] != -1:
        return dp[n]
    
    dp[n] = fib_dp(n-1, dp) + fib_dp(n-2, dp)
    return dp[n]


n = 5
dp = [-1] * (n + 1)

print(fib_dp(n, dp))
```
