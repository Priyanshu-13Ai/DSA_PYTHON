*What is Recursion?*

Recursion tab hoti hai jab ek function apne aap ko hi call karta hai
jab tak condition false na ho jaye.

Recursion = Function calling itself to solve a smaller version of the same problem

______________________________________________________________________________________

Structure of Recursion 

Har recursion me 2 cheezein compulsory hoti hain 

1️. Base Case

Jaha recursion rukta hai

Agar base case nahi → infinite loop 

2️. Recursive Call

Function khud ko smaller problem ke saath call karta hai

_________________________________________________________________________________________

Types of Recursion

1. Direct Recursion

Function directly khud ko call karta hai

EX. def fun():
    print("Hello")
    fun()

2. Indirect Recursion

Ek function doosre function ko call kare
aur wo phir pehle function ko call kare

EX. def fun1():
    fun2()

def fun2():
    fun1()

3. Tail Recursion

Recursive call last statement hoti hai

EX. def fact(n, ans=1):
    if n == 0:
        return ans
    return fact(n-1, ans*n)
Print pehle hota hai

4. Head Recursion

Recursive call pehle hoti hai, baad me kaam

EX. def print_num(n):
    if n == 0:
        return
    print_num(n-1)
    print(n)
print bad me hota hai

5. Tree Recursion

Function multiple recursive calls karta hai

EX. def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

6. Nested Recursion

Recursive call ke andar recursive call

EX. def fun(n):
    if n > 100:
        return n
    return fun(fun(n+11))

```python
# 1. print numbers n to 1
def number(n):
    if n == 0:          # base case
        return           #Jab n = 0 ho jaye → function ruk jaye 
                         #Agar ye nahi likhte ❌ → infinite recursion → error
    print(n)
    number(n-1)         # recursive call
number(5)
```
