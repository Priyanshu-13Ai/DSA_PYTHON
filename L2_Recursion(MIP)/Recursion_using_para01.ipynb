*Recursion Using Parameters*

```python
#print x--> n times
def func(x,n):
    if n==0:
        return 
    print(x)
    func(x,n-1) # agar n karoge to wo chalte jayega n-n-n-n

func(15,4)
```

```python
# Que--> print 1 to n using recursion
def fun(i,n):
    if i>n:
        return
    print(i)
    fun(i+1,n) # i ki value increase kari humne 
fun(1,4)
```

```python
# (ye n to 1 hoga is baar)
def fun(i,n):
    if i>n:
        return
    fun(i+1,n) # i ki value increase kari humne 
    print(i)
fun(1,4)
```

```python
# n to 1 
def fun(n):
    if n==0:
        return
    print(n)
    fun(n-1)
fun(4)
```

```python
# print 1 to n using Tail
def fun(i,n):
    if i==n:
     return
    fun(i+1,n)
    print(i)
fun(1,4)
```
