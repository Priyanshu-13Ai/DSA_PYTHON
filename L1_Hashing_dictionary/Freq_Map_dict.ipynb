Frequency_map

```python
nums =[5,6,7,7,1,9,111,1,1,5,1,1]
# 5:2 ,7:2 aise dictionary honi cahia
```

```python
#Method 1 
nums =[5,6,7,7,1,9,111,1,1,5,1,1]
freq_map={}
for i in range(0,len(nums)):
    if nums[i] in freq_map:
       freq_map[nums[i]]+=1
    else:
        freq_map[nums[i]]=1
print(freq_map)
```

```python
# Method 2
nums =[5,6,7,7,1,9,111,1,1,5,1,1]
hash_map={}
n=len(nums) #List ka total size nikal liya
for i in range(0,n):
    hash_map[nums[i]]=hash_map.get(nums[i],0)+1 
    # get() check karta hai:
    # agar key present hai → uski value return
    # agar key present nahi hai → 0 return
print(hash_map)
```

    HASHING IN PYTHON

```python
"""
constraints
1) i<=n[i]<=10
2) n can have 10^8 elements
3) m can have 10^8 elements
find how many time element of m occcure in n
n =[5,3,2,2,1,5,5,7,5,10]
m=[10,111,1,8,5,67,2]
"""
```

```python
n =[5,3,2,2,1,5,5,7,5,10]
m=[10,111,1,8,5,67,2]
hash_list = [0]*11 # 11 length ki list jisme har index 0 se initilizee hai

for num in n:
    hash_list[num]+=1   # hash_list ke indexes ke andar no of
                        # occurence count ho gai
for num in m:
    if num<1 or num>10:
        print(0)
    else:
      print(hash_list[num])

# Time Complexity O(n+m)
# Space Complexity O(11) --> O(1)
```

Character Hashing

```python
''' que) s me har character kitni baar aaya hai, wo count karna
aur phir q ke har character ka count print karna'''
#constraints
# 1)'a'<=s[i]<='z
# isme ascii code use lena hai 
s = "azyxyyzaaaa"
q = ['d','a','y','x']
hash_list = [0]*26
for ch in s:
    ascii_val = ord(ch)# ord() kisi character ka ASCII / Unicode number batata hai.
    index = ascii_val - 97
    hash_list[index]+=1
for ch in q:
    ascii_val = ord(ch)
    index = ascii_val-97
    print(hash_list[index])

# Time Complexity O(n+m)
# Space Complexity o(26)-->O(1)

```

```python
s = "azyxyyzaaaa"

freq = {}   # empty dictionary

for ch in s:
    # agar character pehle se hai
    if ch in freq:
        freq[ch] += 1
    else:
        # agar pehli baar aaya
        freq[ch] = 1

print(freq)
```

```python
# Given string jisme characters ka frequency count karna hai
s = "azyxyyzaaaa"

# Query list: in characters ka count chahiye
q = ['d', 'a', 'y', 'x']

# Empty dictionary banayi frequency store karne ke liye
freq = {}

# String ke har character par loop
for ch in s:
    # get(ch, 0) ka matlab:
    # agar ch dictionary me hai → uski value
    # nahi hai → 0 return karega
    freq[ch] = freq.get(ch, 0) + 1

# Ab query characters ka count print karenge
for ch in q:
    # agar character dictionary me nahi hoga (jaise 'd')
    # to 0 print hoga
    print(freq.get(ch, 0))
```

```python
s = "azyxyyzaaaa"
q = ['d', 'a', 'y', 'x']

# Empty dictionary for frequency
freq = {}

# Frequency count using dictionary
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

# Print in "character:count" format
for ch in q:
    print(f"{ch}:{freq.get(ch, 0)}")
```
